# Description Dataset Hotel Booking
Dataset ini merupakan data pemesanan hotel dari tgl 1 Juli 2015 - 31 Agustus 2017  di City Hotel dan Resort Hotel. Isi dari dataset terbut termasuk dengan pemsanan yang berhasil maupun yang dibatalkan.

| No. | Nama Kolom | Penjelasan |
| :- | :- | :- |
1| hotel | Berisi jenis hotel yaitu City Hotel dan Resort Hotel |
2| is_canceled | Status pembatalan |
3| lead_time | Jumlah hari yang berlalu antara tanggal masuk pemesanan dan tanggal kedatangan |
4| arrival_date_year | Tahun dari tanggal kedatangan |
5| arrival_date_month | Bulan dari tanggal kedatangan |
6| arrival_date_week_number | Minggu ke- dari tanggal kedatangan |
7| arrival_date_day_of_month | Hari dari bulan tanggal kedatangan |
8| stays_in_weekend_nights  | Jumlah dari pelanggan memesan atau tinggal di hotel pada malam akhir pekan (sabtu malam atau minggu malam) |
9| stays_in_week_nights | Jumlah dari pelanggan memesan atau tinggal di hotel pada malam hari kerja |
10| adults | Jumlah orang dewasa |
11| children | Jumlah anak-anak |
12| babies | Jumlah bayi |
13| meal | Status pelanggan memesan paket makan |
14| country | Negara asal |
15| market_segment | Market segment berupa TA (Travel Agents), TO (Tour Operators) dan lainnya |
16| distribution_channel | Saluran distribusi pemesanan berupa TA (Travel Agents), TO (Tour Operators) dan lainnya |
17| is_repeated_guest | Status apakah pelanggan pernah memesan hotel sebelumnya |
18| previous_cancellations | Jumlah pemesanan hotel yang dibatalkan sebelum pemesanan ini |
19| previous_bookings_not_canceled | Jumlah pemesaran hotel yang tidak dibatalkan sebelum pemesanan ini |
20| reserved_room_type | Kode tipe kamar yang dipesan |
21| assigned_room_type | Kode tipe kamar yang ditetapkan |
22| booking_changes | JUmlah perubahan pada pemesanan yang dilakukan |
23| deposit_type | Tipe deposit (No Deposit - Non Refund - Refundable - Others) |
24| agent | ID dari travel agen yang melakukan pemesanan |
25| company | ID dari perusahaan yang melakukan pemesanan |
26| days_in_waiting_list | Jumlah hari pemesanan dalam daftar tunggu sebelum dikonfirmasi ke pelanggan |
27| customer_type | Tipe pelanggan (Group - Transient - Transient-Party)
28| adr | Average Daily Rate |
29| required_car_parking_spaces | Juimlah tempat parkir mobil yang dibutuhkan |
30| total_of_special_requests | Jumlah permintaan khusus yang dibuat oleh pelanggan |
31| reservation_status | Status pemesanan (Check-Out - No-Show - Canceled)
32| reservation_status_date | Tanggal dari status pemesanan terakhir |
33| name | Nama pelanggan |
34| email | Email pelanggan |
35| phone-number | Nomor telepon pelanggan |
36| credit_card | Nomor kartu kredit pelanggan |

### Total Canceled Booking Hotel Rate 

```python
data_cancel = df[df['is_canceled'] == 1]
data_cancel 
```

```python
rate_cancel_month_year = df.groupby(['arrival_date_year', 'arrival_date_month'])['is_canceled'].count().reset_index()
rate_cancel_month_year
```

```python
adr_month_year = df.groupby(['arrival_date_year', 'arrival_date_month'])['adr'].sum().reset_index()
adr_month_year
```

```python
# jumlah pelanggan yang cancel di grup per bulan
cancel_month_year = data_cancel.groupby(['arrival_date_year', 'arrival_date_month'])['is_canceled'].sum().reset_index()
cancel_month_year
```

```python
cancel_month_year['not_canceled'] = rate_cancel_month_year['is_canceled'] - cancel_month_year['is_canceled'] 
cancel_month_year['perc'] = cancel_month_year['is_canceled'] / rate_cancel_month_year['is_canceled'] * 100


cancel_month_year
```

```python
cancel_month_year['perc'].describe()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x='arrival_date_month', y='perc',data=cancel_month_year,
            hue='arrival_date_year', order=hotel_month_order, palette='icefire')
plt.title('Persentase Customer Cancel')
plt.tight_layout()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x='arrival_date_month', y='not_canceled',data=cancel_month_year,
            hue='arrival_date_year', order=hotel_month_order, palette='icefire')
plt.title('jumlah not_canceled booking')
plt.tight_layout()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x='arrival_date_month', y='adr',data=adr_month_year,
            hue='arrival_date_year', order=hotel_month_order, palette='icefire')
plt.title('adr perbulan')
plt.tight_layout()
```

### Business Insight

```python
hotel_year_month = df[['is_canceled','adr', 'hotel','arrival_date_month', 'arrival_date_year']]

hotel_year_month['date'] = hotel_year_month['arrival_date_year'].astype(str) + '-' + hotel_year_month['arrival_date_month'] 
hotel_year_month['date'] = pd.to_datetime(hotel_year_month['date'])

hotel_year_month['stats'] = hotel_year_month['is_canceled'].apply(lambda x: 'Cancel' if x==1 else 'Not Cancel')
hotel_year_month
```

```python
stats_cancel_date_agg = hotel_year_month.groupby(['date', 'stats'])['is_canceled'].count().reset_index()
stats_cancel_date_agg
```

```python
plt.figure(figsize=(15,8))

sns.lineplot(x='date', y='is_canceled', data=stats_cancel_date_agg,
             hue='stats', marker="o", linewidth=3, linestyle='-', palette='Set1')

plt.axvline(x = pd.to_datetime('2015-12-01'), color='green', linestyle='-.')
plt.axvline(x = pd.to_datetime('2016-12-01'), color='green', linestyle='-.')

plt.xlim(pd.to_datetime('2015-07-01'), pd.to_datetime('2017-08-01'))
plt.xticks(ticks=stats_cancel_date_agg['date'].tolist(), rotation=45)

plt.title('Status Pemesanan Hotel', weight='bold', fontsize=16)
plt.xlabel('Tanggal')
plt.ylabel('Jumlah pemesanan')

# plt.grid()
plt.tight_layout()
```

```python
stats_adr_date_agg = hotel_year_month.groupby(['date', 'stats'])['adr'].sum().reset_index()
stats_adr_date_agg
```

```python
plt.figure(figsize=(15,8))

sns.lineplot(x='date', y='adr', data=stats_adr_date_agg,
             hue='stats', marker="o", linewidth=3, linestyle='-', palette='Set1')

plt.axvline(x = pd.to_datetime('2015-12-01'), color='green', linestyle='-.')
plt.axvline(x = pd.to_datetime('2016-12-01'), color='green', linestyle='-.')

plt.xlim(pd.to_datetime('2015-07-01'), pd.to_datetime('2017-08-01'))
plt.xticks(ticks=stats_adr_date_agg['date'].tolist(), rotation=45)

plt.title('Total ADR', weight='bold', fontsize=16)
plt.xlabel('Tanggal')
plt.ylabel('Total ADR')

# plt.grid()
plt.tight_layout()
```

```python
stats_hotel_date_agg = hotel_year_month.groupby(['date', 'hotel'])['adr'].sum().reset_index()
stats_hotel_date_agg
```

```python
plt.figure(figsize=(15,8))

sns.lineplot(x='date', y='adr', data=stats_hotel_date_agg,
             hue='hotel', marker="o", linewidth=3, linestyle='-', palette='Set1')

plt.axvline(x = pd.to_datetime('2015-12-01'), color='green', linestyle='-.')
plt.axvline(x = pd.to_datetime('2016-12-01'), color='green', linestyle='-.')

plt.xlim(pd.to_datetime('2015-07-01'), pd.to_datetime('2017-08-01'))
plt.xticks(ticks=stats_hotel_date_agg['date'].tolist(), rotation=45)

plt.title('Room price', weight='bold', fontsize=16)
plt.xlabel('Tanggal')
plt.ylabel('Total ADR')

plt.grid()
plt.tight_layout()
```
